```python
# Removes Document Identifier, Removes Topic Label, Returns List of Labels and Documents
def clean_data(file_name):
    with open(file_name, "r") as f:
        all_docs = []
        all_labels = []
        for line in f:
            line_list = list(line.split(" "))
            line_list.pop(0)
            line_list.pop(1)
            line_list.pop(len(line_list)-1)
            all_labels.append(line_list.pop(0))
            all_docs.append(" ".join(line_list))
        return all_docs,all_labels
                

# Takes in documents, labels and returns three lists corresponding to training, validation and test data an labels
def tvt_split_data(all_docs, all_labels):
    test_split_point = int(0.80 * len(all_docs))
    test_data = [all_docs[test_split_point:], all_labels[test_split_point:]]
    training_split_point = int(.80 * len(all_docs[:test_split_point])) 
    training_data = [all_docs[:training_split_point], all_labels[:training_split_point]]
    validation_data = [all_docs[training_split_point:test_split_point], all_labels[training_split_point:test_split_point]]
    return training_data, validation_data, test_data

# Takes in documents, labels and returns two lists corresponding to training and test data an labels
def tt_split_data(all_docs, all_labels):
    split_point = int(0.80 * len(all_docs))
    test_data = [all_docs[split_point:], all_labels[split_point:]]
    training_data = [all_docs[:split_point], all_labels[:split_point]]
    return training_data, test_data

# Encodes Labels
from sklearn import preprocessing
import numpy as np
def encode_labels(labels):
    label_encoder = preprocessing.LabelEncoder()
    return list(label_encoder.fit_transform(labels))

#Generate Performance Metrics

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


def print_confusion_matrix(true_labels, predicted_labels, file_name):
    cm = confusion_matrix(true_labels, predicted_labels)#.ravel()
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["neg", "pos"])
    fig = disp.plot().figure_ 
    fig.savefig(file_name)
    with open("my_file.txt", 'w') as f:
        f.write(np.array2string(cm, separator=', '))
    
    
def print_metrics(true_labels, predicted_labels):  
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels)
    recall = recall_score(true_labels, predicted_labels)
    f1 = f1_score(true_labels, predicted_labels)
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1: {f1}")
    print(f"Accuracy: {accuracy}")
    
def print_predictions(predicted_labels, start_line):
    for i in range(0, len(predicted_labels)):
        print(f"{start_line + i},{predicted_labels[i]}")
        
def print_output(true_labels, predicted_labels, start_line, file_name):
    cm = confusion_matrix(true_labels, predicted_labels)#.ravel()
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["neg", "pos"])
    fig = disp.plot().figure_ 
    fig.savefig("cm.png")    
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels)
    recall = recall_score(true_labels, predicted_labels)
    f1 = f1_score(true_labels, predicted_labels)
    f = open(file_name, "a")
    for i in range(0, len(predicted_labels)):
        f.write(f"{start_line + i},{predicted_labels[i]}\n") 
    f.write("\nConfusion Matrix\n")
    f.write(np.array2string(cm, separator=', '))
    f.write("\n")   
    f.write(f"\nPrecision: {precision}\n")
    f.write(f"Recall: {recall}\n")
    f.write(f"F1: {f1}\n")
    f.write(f"\nAccuracy: {accuracy}\n")
    f.flush()
    f.close()
```

```python
all_docs, all_labels = clean_data("all_sentiment_shuffled.txt")
training, validation, test = tvt_split_data(all_docs, all_labels)
print(len(training[0]))
print(len(validation[0]))
print(len(test[0]))
print(len(all_docs))

training_lines_nos=[0 + 1, len(training[0])]
print(training_lines_nos)
validation_lines_nos = [training_lines_nos[1] + 1, training_lines_nos[1] + len(validation[0])]
print(validation_lines_nos)
test_lines_nos = [validation_lines_nos[1] + 1, validation_lines_nos[1] + len(test[0])]
print(test_lines_nos)

training[1] = encode_labels(training[1])
validation[1] = encode_labels(validation[1])
test[1] = encode_labels(test[1])
```

```python
#Feature Extraction (TFIDF)
from sklearn.feature_extraction.text import TfidfVectorizer
tvec = TfidfVectorizer()
tfidf_training = [tvec.fit_transform(training[0]).toarray(), training[1]]
tfidf_validation = [tvec.transform(validation[0]).toarray(), validation[1]]
tfidf_test = [tvec.transform(test[0]).toarray(), test[1]]                                                      


#Feature Extraction (Count Vector)
from sklearn.feature_extraction.text import CountVectorizer
cvec = CountVectorizer()
count_training = [cvec.fit_transform(training[0]).toarray(), training[1]]
count_validation = [cvec.transform(validation[0]).toarray(), validation[1]]
count_test = [cvec.transform(test[0]).toarray(), test[1]]  
```

```python
#Run Naive Bayes Classifier using default smoothing = 1, using feature vectors generated using a Count Vector with default parameters(casefolding, no stop words removed)
import joblib
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB() 
mnb.fit(count_training[0], count_training[1])
predicted_labels = mnb.predict(count_validation[0])
print_confusion_matrix(count_validation[1], predicted_labels, "nbayes_count_no_smoothing.png")
print_metrics(count_validation[1], predicted_labels)
```

```python
#Optimizing for alpha 
import joblib
from scipy.stats import uniform
from sklearn.model_selection import RandomizedSearchCV
space = {'alpha': uniform(0,1)}
optimized_mnb = RandomizedSearchCV(mnb, space, n_iter=50, scoring='accuracy', random_state=0)
result = optimized_mnb.fit(count_validation[0], count_validation[1])
print(result.best_params_)
joblib.dump(result, "optimized_nbayes_count_smoothing.pkl")
predicted_labels = optimized_mnb.predict(count_validation[0])
print_confusion_matrix(count_validation[1], predicted_labels, "optimized_nbayes_count_smoothing.png")
print_metrics(count_validation[1], predicted_labels)
```

```python
#Test set
import joblib
optimized_mnb = joblib.load("optimized_nbayes_count_smoothing.pkl")
predicted_labels = optimized_mnb.predict(count_test[0])
predicted_probabilities = optimized_mnb.predict_proba(count_test[0])
print_confusion_matrix(count_test[1], predicted_labels, "optimized_nbayes_count-test.png")
print_metrics(count_test[1], predicted_labels)
print_output(count_test[1], predicted_labels, test_lines_nos[0], "optimized_nbayes_count-test.txt")
```

```python
import joblib
correct_indexes = []
incorrect_indexes = []
for i in range(0, len(count_test[1])):
    if count_test[1][i] != predicted_labels[i]:
        incorrect_indexes.append(i)
    else:
        correct_indexes.append(i)

incorrect_documents = []
for i in incorrect_indexes:
    incorrect_documents.append([all_docs[i + test_lines_nos[0] - 1], all_labels[i + test_lines_nos[0] - 1], predicted_probabilities[i]])

correct_documents = []
for i in correct_indexes:
    correct_documents.append([all_docs[i + test_lines_nos[0] - 1], all_labels[i + test_lines_nos[0] - 1], predicted_probabilities[i]])

joblib.dump(incorrect_documents, 'nb_incorrect.pkl')
joblib.dump(correct_documents, 'nb_correct.pkl')
```
